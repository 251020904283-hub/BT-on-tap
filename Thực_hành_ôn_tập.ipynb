```
# Bài 1
gpa = float(input("Nhập điểm GPA: "))
if gpa < 0 or gpa > 10:
    print("Điểm GPA không hợp lệ")
elif gpa >= 8.5 or gpa <= 10:
    print("Học lực: Xuất sắc")
elif gpa >= 7.0:
    print("Học lực: Giỏi")
elif gpa >= 5.5:
    print("Học lực: Khá")
elif gpa >= 4.0:
    print("Học lực: Trung bình")
else:
    print("Học lực: Yếu")
```

```
# Bài 2
n = int(input("Nhập số n: "))
tong = 0
for i in range(1, n + 1):
    if i % 2 == 0:
        tong += i
print("Tổng các số chẵn từ 1 đến", n, "là:", tong)
```

```
# Bài 3
mat_khau_dung = "python123"
so_lan_nhap = 0
toi_da = 3

while so_lan_nhap < toi_da:
    mat_khau = input("Nhập mật khẩu: ")
    so_lan_nhap += 1

    if mat_khau == mat_khau_dung:
        print("Đăng nhập thành công")
        break
else:
    print("Tài khoản bị khóa")
```
